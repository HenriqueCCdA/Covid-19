```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('casos_e_mortes_confirmados.csv')
```

```python
dados_grupo = dados.groupby('Pais')
```

```python
dados_grupo.get_group('Brasil').tail(n=5)
```

```python
# divisao do eixo x
x_ticks = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55]
```

## <span style='color:green;font-size:150%'> Plotando o gráfico</span>
---

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))

name = 'Brasil'
ax1.plot(dados_grupo.get_group(name).Dias, 
             dados_grupo['Casos_Media_10'].get_group(name), 
             label = 'Média movel de 10 dias', 
             color = 'black',
             ls    = '-')

ax1.bar(dados_grupo.get_group(name).Dias, 
        dados_grupo['Casos'].get_group(name),
        label='Acumulado',
        alpha=0.5)

ax1.plot(dados_grupo.get_group(name).Dias, 
             dados_grupo['Taxa_Casos_Media_10'].get_group(name), 
             color = 'black',
             ls    = '-')

ax1.bar(dados_grupo.get_group(name).Dias, 
        dados_grupo['Taxa_Casos'].get_group(name),
        color = 'r',
        label='Novos',
        alpha=0.5)
#

ax2.plot(dados_grupo.get_group(name).Dias, 
             dados_grupo['Mortes_Media_10'].get_group(name), 
             color = 'black',
             ls    = '-')

ax2.bar(dados_grupo.get_group(name).Dias, 
        dados_grupo['Mortes'].get_group(name),
        alpha=0.5)

ax2.plot(dados_grupo.get_group(name).Dias, 
             dados_grupo['Taxa_Mortes_Media_10'].get_group(name), 
             label = 'Média movel de 10 dias', 
             color = 'black',
             ls    = '-')

ax2.bar(dados_grupo.get_group(name).Dias, 
        dados_grupo['Taxa_Mortes'].get_group(name),
        color = 'r',
        alpha=0.5)


# eixo 1
    
# Set the tick labels font
for label in (ax1.get_xticklabels() + ax1.get_yticklabels()):
    label.set_fontsize(16)
#ax1.set_yscale('log')
ax1.set_ylim(0.1, 1500000)
ax1.set_xlim(0, 100)
ax1.set_xlabel('Número de dias apartir do 200° caso', fontsize = 16)
ax1.set_ylabel('Número de Casos', fontsize = 16)
ax1.legend(loc = 'upper left', fontsize = 14)
ax1.set_title('Casos Totais Confirmados', fontsize = 28)
ax1.grid(ls = '--')

# eixo 2

# Set the tick labels font
for label in (ax2.get_xticklabels() + ax2.get_yticklabels()):
    label.set_fontsize(16)

#ax2.set_yscale('log') 
ax2.set_ylim(0.1, 60000)
ax2.set_xlim(0, 100)
ax2.set_xlabel('Número de dias apartir do 200° caso', fontsize = 16)
ax2.set_ylabel('Número de Mortes confirmadas', fontsize = 16)
ax2.set_title('Mortes confirmadas', fontsize = 28)
ax2.grid(ls = '--')

plt.show()
```

```python
fig.savefig('brasil.png', dpi = 300)
```
