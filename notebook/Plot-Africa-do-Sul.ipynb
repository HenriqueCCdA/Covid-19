```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('casos_e_mortes_confirmados.csv')
```

```python
dados_grupo = dados.groupby('Pais')
```

```python
dados_grupo.get_group('África do Sul').tail()
```

```python
paises = ['África do Sul', 'Coreia do Sul', 'USA']
```

____
____
## <span style='color:green;font-size:150%'> Plotando o gráfico dos casos totais confirmados</span>
____

```python
x_ticks = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55]
m       = ['o', 'v', '^', 'x', 's', '>']

fig, ax = plt.subplots(figsize=(16, 6))

for i, name in enumerate(paises):
    ax.semilogy(dados_grupo.get_group(name).Dias, dados_grupo['Casos_Media_3'].get_group(name), label = name, marker = m[i])

# Set the tick labels font
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(16)

ax.set_xlim(0, 55)
ax.set_xlabel('Número de dias apartir do 200° caso', fontsize = 18)
ax.set_ylabel('Número de Casos', fontsize = 18)
ax.legend(loc = 'lower right', fontsize = 20)
ax.set_title('Casos', fontsize = 28)
ax.grid(ls = '--')
ax.axvline(x=4.5,ls='--',color='red')
ax.text(2,1200, 'Início do isolamento na\nAfrica do Sul', fontsize=12, rotation=90, color = 'red')
ax.text(2,95, '24/03 : 00:00 h', fontsize=12, rotation=0, color = 'red')
ax
```
